```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.model_selection import train_test_split
SEED = 28
df = pd.read_csv("train.csv")
df.columns = [col.lower() for col in df.columns]
```

```python
df
```

```python
def one_hot_encoding(df,category):
    for name in list(df[f'{category}'].unique()):
        df[f'{category}_{name}'] = df[f'{category}'].map(lambda x:1 if x==name else 0)
    df.drop(category, axis=1, inplace=True)
```

```python
df
```

```python
# df['gender'] = df['gender'].map(lambda x:1 if x=='male' else 2)

# def embarkCheck(value):
#     if value=='S': return 1
#     elif value=='C': return 2
#     else: return 3
# df['embarked'] = df['embarked'].map(lambda x:embarkCheck(x))


# X = df.drop('survived', axis=1)
X = df.drop(['survived','ticket','cabin','name'], axis=1)
y = df['survived']
X_tr, X_te, y_tr, y_te = train_test_split(X, y, random_state=SEED, test_size=0.2)

# SEED = 42

# X_tr, X_te = train_test_split(df, random_state=SEED, test_size = 0.2)
# X_tr = X_tr.reset_index(drop=True)
# X_te = X_te.reset_index(drop=True)

X_tr.shape, X_te.shape

X_tr = X_tr.reset_index(drop=True)
X_te = X_te.reset_index(drop=True)

y_tr = y_tr.reset_index(drop=True)
y_te = y_te.reset_index(drop=True)
```

```python
dict_designation = {
    'Mr.': '남성',
    'Master.': '남성',
    'Sir.': '남성',
    'Miss.': '미혼 여성',
    'Mrs.': '기혼 여성',
    'Ms.': '미혼/기혼 여성',
    'Lady.': '숙녀',
    'Mlle.': '아가씨',
    # 직업
    'Dr.': '의사',
    'Rev.': '목사',
    'Major.': '계급',
    'Don.': '교수',
    'Col.': '군인',
    'Capt.': '군인',
    # 귀족
    'Mme.': '영부인',
    'Countess.': '백작부인',
    'Jonkheer.': '귀족'
}

def add_designation(name): # 호칭 함수
    designation = "unknown"
    for key in dict_designation.keys():
        if key in name:
            designation = key
            break
    return designation

X_tr['designation'] = X_tr['name'].map(lambda x: add_designation(x))
X_te['designation'] = X_te['name'].map(lambda x: add_designation(x))

X_tr.head()
```

```python
X_tr['age'] = X_tr['age'].fillna(X_tr['age'].median())
X_te['age'] = X_te['age'].fillna(X_tr['age'].median())

X_tr['fare'] = X_tr['fare'].fillna(X_tr['fare'].median())
X_te['fare'] = X_te['fare'].fillna(X_tr['fare'].median())

em_mode = X_tr['embarked'].mode().values[0]
X_tr['embarked'] = X_tr['embarked'].fillna(em_mode)
X_te['embarked'] = X_te['embarked'].fillna(em_mode)

# y_tr['age'] = y_tr['age'].fillna(y_tr['age'].median())
# y_te['age'] = y_te['age'].fillna(y_te['age'].median())
```

```python
X_tr['fare'] = X_tr['fare'].map(lambda i: np.log(i) if i > 0 else 0)
X_te['fare'] = X_te['fare'].map(lambda i : np.log(i) if i > 0 else 0)
```

```python
# survived
X_tr["age"] = X_tr["age"].astype("int32")
X_te["age"] = X_te["age"].astype("int32")

# X_tr["fare"] = X_tr["fare"].astype("int32")
# X_te["fare"] = X_te["fare"].astype("int32")
```

```python
X_te.drop('name',axis=1,inplace=True)
```

```python
X_tr['f_size'] = X_tr['sibsp']+X_tr['parch']
X_te['f_size'] = X_te['sibsp']+X_te['parch']
```

```python
X_tr.drop(['sibsp','parch'],axis=1,inplace=True)
X_te.drop(['sibsp','parch'],axis=1,inplace=True)
```

```python
X_tr['f_size'].unique()
```

```python
# def family_size(value):
#     if value==0: return 0
#     elif value==1: return 1
#     elif value<5: return 2
#     else: return 3
def family_size(value):
    if value<=1: return 1
    else: return 2
X_tr['f_size'] = X_tr['f_size'].map(lambda x:family_size(x))
X_te['f_size'] = X_te['f_size'].map(lambda x:family_size(x))
```
