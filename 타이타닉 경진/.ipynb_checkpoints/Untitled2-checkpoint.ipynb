```python
# 'Mr', 'Master', 'Mrs', 'Miss'이외는 other로 저장
def clean_designation(title):
    if title in ['Mr', 'Miss', 'Mrs', 'Master']:
        return title
    else:
        return 'other'

titanic_train['designation']=titanic_train['designation'].apply(clean_designation)
titanic_test['designation']=titanic_test['designation'].apply(clean_designation)
```

```python
def one_hot_encoding(df,category):
    for name in list(df[f'{category}'].unique()):
        df[f'{category}_{name}'] = df[f'{category}'].map(lambda x:1 if x==name else 0)
    df.drop(category, axis=1, inplace=True)
```
