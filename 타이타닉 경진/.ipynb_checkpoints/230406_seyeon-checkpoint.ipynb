##Missing Value: filling

Missing Value를 여기에다 둔 이유는 결측값의 존재로 인해 원활한 피쳐 추출에 지장이 있기 때문이다. 

```python
train.isnull().sum()
```

```python
test.isnull().sum()
```

```python
ori_te.isnull().sum()
```

```python
age_median = train['age'].median()
fare_median = train['fare'].median()
embarked_mode = train['embarked'].mode().values[0]

age_median, fare_median, embarked_mode
```

```python
train['age'].fillna(age_median, inplace=True)
test['age'].fillna(age_median, inplace=True)
ori_te['age'].fillna(age_median, inplace=True)

train['fare'].fillna(fare_median, inplace=True)
test['fare'].fillna(fare_median, inplace=True)
ori_te['fare'].fillna(fare_median, inplace=True)

train['embarked'].fillna(embarked_mode, inplace=True)
test['embarked'].fillna(embarked_mode, inplace=True)
ori_te['embarked'].fillna(embarked_mode, inplace=True)

train.isnull().sum().sum(), test.isnull().sum().sum(), ori_te.isnull().sum().sum()
```

##Function Definition

```python
# 이산형/범주형 데이터 category plot 그리기

def cat_plot(feature1, hue):
  sns.catplot(data=train, x=feature1, y="survived", hue=hue, kind="bar")
```

```python
# 범주형 데이터 bar chart 그리기
def bar_chart(feature):
    survived = train[train['survived']==1][feature].value_counts()
    dead = train[train['survived']==0][feature].value_counts()
    df = pd.DataFrame([survived,dead])
    df.index = ['survived','Dead']
    df.plot(kind='bar',stacked=True, figsize=(10,5))
```

```python
# 수치형 데이터 displot 그리기
def dis_plot(feature1, feature2, feature3):
  sns.displot(train, x=feature1, col=feature2, row=feature3, hue="survived", multiple="dodge", bins=20)
```

```python
# train 데이터 상관관계 파악 _ 1.통계치, 2.히트맵 

def correlation_heatmap(df):
    _ , ax = plt.subplots(figsize =(14, 12))
    colormap = sns.diverging_palette(220, 10, as_cmap = True)
    
    _ = sns.heatmap(
        df.corr(), 
        cmap = colormap,
        square=True, 
        cbar_kws={'shrink':.9 }, 
        ax=ax,
        annot=True, 
        linewidths=0.1,vmax=1.0, linecolor='white',
        annot_kws={'fontsize':12 }
    )
    
    plt.title('Pearson Correlation of Features', y=1.05, size=15)

print(train.corr())
correlation_heatmap(train)
```

#####infant 

    age <= 4

    dtype: int (0,1)

```python
for i in range(1,10,1):
  print(f'age가 {i}이하일 때 평균생존율:', train[train['age'] < i]['survived'].mean().round(3))
```

```python
def add_infant(age):
  result = 0
  try:
    if age <= 4:
      result = 1     
  except:
    pass
  return result

train['infant'] = train['age'].map(lambda x: add_infant(x))
test['infant'] = test['age'].map(lambda x: add_infant(x))
ori_te['infant'] = ori_te['age'].map(lambda x: add_infant(x))


train.columns
```

#####young_male

    조건: 
    1. gender == "male"
    2. 12<=age<=24

    dtype: int (0,1)

```python
# 일단 그래프에서 눈대중으로 밴드 잡아놓고 생존율 계산 시작
for i in range(4,15,1):
  for j in range(19,25,1):
    print(f'청년남성 age band가 {i}~{j}일 때 평균생존율:', train[(train['age']>=i)&(train['age']<=j)&(train['gender']=='male')]['survived'].mean().round(3))
```

```python
def young_male(cols):
  result = 0 
  try:
    if (cols['age'] >= 4 ) & (cols['age'] <= 24) & (cols['gender']=='male'):
      result = 1
  except:
    pass
  return result

train['young_male'] = train.apply(lambda cols: young_male(cols), axis=1)
test['young_male'] = test.apply(lambda cols: young_male(cols), axis=1)
ori_te['young_male'] = ori_te.apply(lambda cols: young_male(cols), axis=1)
```

#####middle_name_cat

    조건:
    1. middle_name을 value_counts() 순으로 4가지로 분류

    dtype: object (Mr, Miss, Mrs, others)

```python
def add_middle_name_others(middle_name):
    if middle_name in ['Miss', 'Mr', 'Mrs']:
        return middle_name
    else:
        return 'others'

train['middle_name_cat'] = train['middle_name'].map(lambda x: add_middle_name_others(x))
test['middle_name_cat'] = test['middle_name'].map(lambda x: add_middle_name_others(x))
ori_te['middle_name_cat'] = ori_te['middle_name'].map(lambda x: add_middle_name_others(x))

train['middle_name_cat'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder 
```

```python
print(f'얼마나 스파스해졌는지 확인\nbefore: {train.shape} / {test.shape}')

enc = OneHotEncoder()
# train :: 문자형 데이터 인코딩 실시
tmp_tr = pd.DataFrame(
    enc.fit_transform(train[enc_cols]).toarray(), 
    columns = enc.get_feature_names_out() 
)
enc_tr = pd.concat(
    [train[normal_cols].reset_index(drop=True), tmp_tr.reset_index(drop=True)]
    , axis=1 
)
# test
tmp_te = pd.DataFrame(
    enc.transform(test[enc_cols]).toarray(), 
    columns = enc.get_feature_names_out()
)
enc_te = pd.concat(
    [test[normal_cols].reset_index(drop=True), tmp_te.reset_index(drop=True)]
    , axis=1
)
# ori_test
tmp_te = pd.DataFrame(
    enc.transform(ori_te[enc_cols]).toarray(), 
    columns = enc.get_feature_names_out()
)
enc_ori_te = pd.concat(
    [ori_te[normal_cols].reset_index(drop=True), tmp_te.reset_index(drop=True)]
    , axis=1
)

print(f'after: {enc_tr.shape} / {enc_te.shape}')
enc_tr.head()
```

confusion matrix를 어떻게 보아야 하는가? 

- 1. test data의 recall, precision의 절대적 수치가 높아야 한다.
- 2. test data의 recall, precision 간 차이가 크지 않아야 한다.
- 3. train data 와 test data 간 recall, precision 차이가 크지 않아야 한다.

만약 상기 사항에서 이상이 있을 경우 어떻게 바로잡을 수 있을지를 고민하기

```python
# Confusion Matrix 
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay

# predict 매서드 사용하여 예측값 뽑기 : modelV2 말고 최적값 숫자 넣어보기
from sklearn.base import BaseEstimator
X_tr_pred = modelV2.predict(enc_tr)
X_te_pred = modelV2.predict(enc_te)

# Confusion Matrix 그리기
fig , ax = plt.subplots(1,2,figsize=(15,5))

cm = confusion_matrix(y_tr,X_tr_pred) #(실제값, 예측값)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(ax = ax[0])

cm = confusion_matrix(y_te,X_te_pred) #(실제값, 예측값)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot(ax = ax[1])
plt.show()

# 통계치 표시
from sklearn.metrics import classification_report
print(classification_report(y_tr, X_tr_pred))
print('-'*100)
print(classification_report(y_te, X_te_pred))
```

train과 test 모두에서 recall이 현저히 낮다.
다시 말해, 실제로는 살았는데 죽었다고 예측하는 경우가 많다는 것.
내가 가진 피쳐 중에서 '이런 사람은 죽었을 가능성이 높다'는 학습을 하게 만드는 피쳐를 더욱 세분화해서 그 중에서 살아있는 경우를 추출해야 한다.


    Feature 정교화
    A = ( 남성, 젊은 사람, fare가 낮은 사람, pclass가 낮은 사람, family_num이 1~3인 사람 ) 
    B = ( middle_name_cat, cabin, embarked ) 
    
    A 중에서 B를 기준으로 나누고, 그 중에서 생존율이 A 카테고리의 평균보다 높은 경우를 추출해 내야겠다.
